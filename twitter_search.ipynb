```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import datetime as dt

# 크롬 드라이버 연결
chromedriver = '/Users/euzl/GitHub/Is-Son-World-Class/document/store/chromedriver'
browser = webdriver.Chrome(chromedriver)

# 키워드, 날짜 설정
startdate = dt.date(year=2020, month=10, day=27)
untildate = dt.date(year=2020, month=10, day=28) # 하루하루의 결과를 출력하기 위한 중간변수
enddate = dt.date(year=2020, month=10, day=28)
keyword = "손흥민"
url='https://twitter.com/search?q='+keyword+'%20since%3A'+str(startdate)+'%20until%3A'+str(untildate)+'&amp;amp;amp;amp;amp;amp;lang=eg'

file = open("/result/twitter_son.txt", 'w')

while not enddate == startdate:
    # url에서 until + <날짜> 는 결과에 포함되지 않는다.
    url='https://twitter.com/search?q='+keyword+'%20since%3A'+str(startdate)+'%20until%3A'+str(untildate)+'&amp;amp;amp;amp;amp;amp;lang=eg'
    
    # 사이트 접속
    browser.get(url)
    html = browser.page_source
    soup=BeautifulSoup(html, 'html.parser')
    
    lastHeight = browser.execute_script("return document.body.scrollHeight")
    
    while True:
        
        tweet = soup.select('div#react-root main section article span')

        if len(tweet) > 20: # 20 이하는 필요없는 데이터 (트위터관련)
            if '@' in tweet:
                    break;
            for x in range(len(tweet)):
                print(tweet[x].get_text(), end=' ')
                file.write(tweet[x].get_text() + ' ')
            print()

        file.write('\n')
        
        # 스크롤 내리는 코드 (트위터 특성상 스크롤을 내려야 한다)
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        time.sleep(1)
        
        newHeight = browser.execute_script("return document.body.scrollHeight")
        
        if newHeight != lastHeight:
            html = browser.page_source
            soup = BeautifulSoup(html, 'html.parser')
        else:
            startdate=untildate
            untildate += dt.timedelta(days=1)
            break
        lastHeight = newHeight
        
file.close()

browser.quit()
```
